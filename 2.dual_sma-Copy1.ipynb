# Стратегия по пересечению двух скользящих средней

Описание стратегии:
<br> Две скользящие средние SMA30 и SMA100.
<br>Если SMA30 пересекает SMA100 сверху, то необходимо продавать. 
<br>Если SMA30 пересекает SMA100 снизу, то необходимо покупать.

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
#Загружаем акцию
stock = 'AAPL'
data = yf.download(stock, '2016-01-01', '2022-08-18')
```

```python
#График по акции
plt.figure(figsize = (13,5))
plt.plot(data['Adj Close'], label = 'Adjusted Close Price')
plt.xlabel('30 Dec 2015 - 17 Aug 2022')
plt.ylabel('Price USD')
plt.legend(loc = 'upper left')
```

```python
#Создаем данные по скользящим
SMA30 = pd.DataFrame()
SMA30['Adj Close'] = data['Adj Close'].rolling(window = 30).mean()
SMA100 = pd.DataFrame()
SMA100['Adj Close'] = data['Adj Close'].rolling(window = 100).mean()
```

```python
#График вместе со скользящими
plt.figure(figsize = (13,5))
plt.plot(data['Adj Close'], label = stock)
plt.plot(SMA30['Adj Close'], label = 'SMA30')
plt.plot(SMA100['Adj Close'], label = 'SMA100')
plt.xlabel('30 Dec 2015 - 17 Aug 2022')
plt.ylabel('Price USD')
plt.legend(loc = 'upper left')
```

```python
df = pd.DataFrame()
df['Adj Close'] = data['Adj Close']
df['SMA30'] = SMA30['Adj Close']
df['SMA100'] = SMA100['Adj Close']
```

```python
def dual_sma(df):
    buy_signal = []
    sell_signal = []
    flag = 0
    
    for i in range(len(df)):
        if df['SMA30'][i] > df['SMA100'][i]:
            if flag != 1:
                buy_signal.append(df['Adj Close'][i])
                sell_signal.append(np.nan)
                flag = 1
                
            else: 
                buy_signal.append(np.nan)
                sell_signal.append(np.nan)
                
        elif df['SMA30'][i] < df['SMA100'][i]:
            if flag != -1:
                sell_signal.append(df['Adj Close'][i])
                buy_signal.append(np.nan)
                flag = -1
                
            else: 
                buy_signal.append(np.nan)
                sell_signal.append(np.nan)
        
        else:
            buy_signal.append(np.nan)
            sell_signal.append(np.nan)
        
    return(buy_signal, sell_signal)
```

```python
dual_sma = dual_sma(df)
df['buy signal'] = dual_sma[0]
df['sell signal'] = dual_sma[1]
```

```python
# График по стратегии
plt.figure(figsize = (13,5))
plt.plot(data['Adj Close'], label = 'Adjusted Close Price', alpha = 0.3, linewidth = 2)
plt.plot(SMA30['Adj Close'], label = 'SMA30', linewidth = 3)
plt.plot(SMA100['Adj Close'], label = 'SMA100', linewidth = 3)
plt.scatter(df.index, df['buy signal'], label = 'buy', color = 'green', marker = '^', linewidth = 5)
plt.scatter(df.index, df['sell signal'], label = 'sell', color = 'red', marker = 'v', linewidth = 5)
plt.xlabel('30 Dec 2015 - 17 Aug 2022')
plt.ylabel('Price USD')
plt.legend(loc = 'upper left')
```
