# Расчет процента акций выше 200-дневной скользящей для DJI

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

Работа по индексу

```python
#Индекс
index = 'DJI'

#Загружаем данные
data = yf.download(index, '2016-01-01', '2022-08-31')

#График по индексу
plt.figure(figsize = (13,5))
plt.plot(data['Adj Close'], label = index, alpha = 0.7, linewidth = 2)
plt.xlabel('30 Dec 2015 - 18 Aug 2022')
plt.ylabel('Price USD')
plt.legend(loc = 'upper left')
```

Работа по компаниям в индексе

```python
#Загружаю состав индекса из списка
index_comp = pd.read_csv('C:/Users/damir/OneDrive/Документы/Трейдинг/Python/Stocks in the Dow Jones Industrial Average.csv')\
    .sort_values('Symbol')

tickers = list(index_comp['Symbol'])

#Загружаем данные по акциям
df = yf.download(tickers, '2016-01-01', '2022-08-31').drop(columns=['Adj Close', 'Volume', 'Low', 'Open'])
```

# Скользящая средняя

```python
#Условие по скользящей
ma_period = 200

#Цикл по расчету 200-дневной скользящей средней
for i in tickers:
    df['MA'+str(ma_period), i] = df['Close', i].rolling(ma_period).mean().fillna(0)  
```

```python
#Расчет кол-ва акций, торгующихся выше своих 200-дневных скользящих
amount = []
percentage = []
for i in range(0, len(df)):
    counter = 0
    for j in tickers:
        if df['Close', j][i] > df['MA'+str(ma_period), j][i]:
            counter += 1
        else:
            counter += 0
    amount.append(counter)
df['Amount'] = amount
df['Percentage'] = df['Amount']/len(tickers)*100
```

```python
start = ma_period
df.drop(index=[start+1], axis = 0)
```

```python
#Визуализация графика

# Создаем график
fig, ax1 = plt.subplots(figsize = (20,10))

ax1.set_xlabel('Date') 
ax1.set_ylabel('Price USD', color = 'red') 
plot_1 = ax1.plot(data['Close'], color = 'red', label= 'Индекс '+str(index), alpha = 0.7, linewidth = 2) 
ax1.tick_params(axis ='y', labelcolor = 'red') 

#Второй график
ax2 = ax1.twinx() 
  
ax2.set_ylabel('Процент акций выше SMA', color = 'green') 
plot_2 = ax2.plot(df['Percentage'], color = 'green', label = 'Процент акций выше SMA'+ str(ma_period), alpha = 0.7, linewidth = 2) 
ax2.tick_params(axis ='y', labelcolor = 'green')     

#Legends
lns = plot_1 + plot_2
labels = [l.get_label() for l in lns]
plt.legend(lns, labels, loc='lower right')

               
plt.show()
```
